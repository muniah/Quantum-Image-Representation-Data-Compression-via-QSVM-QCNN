```python
import matplotlib.pyplot as plt # type: ignore
import numpy as np # type: ignore

# Data Plotting
models = ['SVM', 'QSVM', 'QSVM + INEQR']
train_accuracies = [78.8, 100.0, 93.0]
test_accuracies = [70.0, 50.0, 60.0]
train_times = [0.0246, 22.0, 1.06]
test_times = [0.0117, 25.0, 1.62]

# Bar plot for accuracies
plt.figure(figsize=(8, 6))
bar_width = 0.35
x = np.arange(len(models))

plt.bar(x - bar_width/2, train_accuracies, bar_width, label='Training Accuracy (%)', color='blue', alpha = 0.5)
plt.bar(x + bar_width/2, test_accuracies, bar_width, label='Testing Accuracy (%)', color='red', alpha = 0.5)

# Annotate dataset differences
plt.text(2, 50, "Train size: 15\nTest size: 5\nResolution: 4x4", 
         fontsize=10, color='red', bbox=dict(facecolor='white', alpha=0.8))

# Add accuracy labels
for i in range(len(models)):
    plt.text(x[i] - bar_width/2, train_accuracies[i] + 1, f"{train_accuracies[i]}%", ha='center', color='blue', fontsize=10)
    plt.text(x[i] + bar_width/2, test_accuracies[i] + 1, f"{test_accuracies[i]}%", ha='center', color='red', fontsize=10)

# Plot formatting
plt.xticks(x, models)
plt.title('Model Accuracy Comparison', fontsize=16)
plt.ylabel('Accuracy (%)', fontsize=14)
plt.xlabel('Model', fontsize=14)
plt.legend(fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

```

```python
# Bar plot for training and testing times
plt.figure(figsize=(8, 6))
bar_width = 0.35
plt.bar(x - bar_width/2, train_times, bar_width, label='Training Time (s)', color='purple', log=True, alpha = 0.5)
plt.bar(x + bar_width/2, test_times, bar_width, label='Testing Time (s)', color='orange', log=True, alpha = 0.5)

# Annotate dataset differences
plt.text(2, 0.5, "Train size: 15\nTest size: 5\nResolution: 4x4", 
         fontsize=10, color='red', bbox=dict(facecolor='white', alpha=0.8))

# Add time labels
for i in range(len(models)):
    plt.text(x[i] - bar_width/2, train_times[i] + 0.5, f"{train_times[i]:.2f}s", ha='center', color='purple', fontsize=10, alpha = 0.5)
    plt.text(x[i] + bar_width/2, test_times[i] + 0.5, f"{test_times[i]:.2f}s", ha='center', color='orange', fontsize=10, alpha = 0.5)

# Plot formatting
plt.xticks(x, models)
plt.title('Model Training and Testing Time Comparison', fontsize=16)
plt.ylabel('Time (s)', fontsize=14)
plt.xlabel('Model', fontsize=14)
plt.legend(fontsize=12, loc = "upper left")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt  # type: ignore
import numpy as np  # type: ignore

# Data
models = ['SVM', 'QSVM']
train_times = [0.0246, 22.0]
test_times = [0.0117, 25.0]

# Logarithmic bar plot for training and testing times
plt.figure(figsize=(8, 6))
bar_width = 0.35
x = np.arange(len(models))

plt.bar(x - bar_width/2, train_times, bar_width, label='Training Time (s)', color='orange', log=True, alpha = 0.5)
plt.bar(x + bar_width/2, test_times, bar_width, label='Testing Time (s)', color='green', log=True, alpha = 0.5)

# Add time labels
for i in range(len(models)):
    plt.text(x[i] - bar_width/2, train_times[i] * 1.1, f"{train_times[i]:.2f}s", ha='center', color='orange', fontsize=10)
    plt.text(x[i] + bar_width/2, test_times[i] * 1.1, f"{test_times[i]:.2f}s", ha='center', color='green', fontsize=10)

# Plot formatting
plt.xticks(x, models)
plt.title('SVM vs QSVM: Training and Testing Time', fontsize=14)
plt.ylabel('Time (s, log scale)', fontsize=14)
plt.xlabel('Model', fontsize=14)
plt.legend(fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7, which='both')
plt.tight_layout()
plt.show()
```
